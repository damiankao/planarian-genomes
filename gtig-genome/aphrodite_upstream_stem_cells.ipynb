```python
from collections import defaultdict

prefix = '/home/share/projects/planarian_genome/gtig/misc/'
neoList = dict([(x.split('\t')[0],[x.split('\t')[1],i]) for i,x in enumerate(open(prefix + 'neoblast.data'))])
neoList['OX_Smed_1.0.15112'] = ['NA',-1]
neoList['OX_Smed_1.0.16488'] = ['NA',-1]
neoList['OX_Smed_1.0.16818'] = ['NA',-1]
neoList['OX_Smed_1.0.03200'] = ['NA',-1]
neoList['OX_Smed_1.0.22887'] = ['NA',-1]

blastFile = open(prefix + 'nGt051.faa.OX_smed_1.0.fa.tblastn.1e-5')

topHit = {}
for line in blastFile:
    meta = line.strip().split()
    if not topHit.has_key(meta[0]):
        topHit[meta[0]] = meta[1]
        
topHit = dict([(v,k) for k,v in topHit.items()])

gffFile = open(prefix + 'nGt051.MAKER.gff3')
gtigData = {}
refs = defaultdict(list)
for line in gffFile:
    meta = line.strip().split()
    name = meta[-1].split(';')[0].split('=')[1]
    if meta[2] == 'mRNA':
        ref = meta[0]
        start = int(meta[3])
        end = int(meta[4])
        orient = meta[6]
        gtigData[name] = [ref, start, end, orient]
        refs[ref].append((name, start, end, orient))
        
meanExp = {}
meanExpFile = open(prefix + 'all.deseq.counts')
meanExpFile.next()
for line in meanExpFile:
    meta = line.strip().split()
    vals = [float(x) for x in meta[1:]]
    if neoList.has_key(meta[0]):
        neoList[meta[0]].append(sum(vals) / len(vals))
        if topHit.has_key(meta[0]):
            neoList[meta[0]].append(topHit[meta[0]])
            neoList[meta[0]].append(gtigData[topHit[meta[0]]])
        else:
            neoList[meta[0]].append('No G.Tig ortholog found')
            neoList[meta[0]].append('NA')
```

```python
sortNeo = neoList.items()
sortNeo.sort(key = lambda x : x[1][2], reverse = True)
```

```python
refs = {}
for k,v in neoList.items():
    if v[-1] != 'NA':
        refs[v[-1][0]] = ''
        
refFile = open(prefix + 'nGt05.fasta')

for header in refFile:
    header = header.strip()[1:]
    seq = refFile.next().strip()
    if refs.has_key(header):
        refs[header] = seq
```

```python
def revcomp(s):
    s = s[::-1].upper()
    comp = {'A':'T','T':'A','C':'G','G':'C','N':'N'}
    out = ''
    for b in s:
        out += comp[b]
    return out
   
upOut = ''
downOut = ''
for k,v in neoList.items():
    if v[-1] != 'NA':
        ref = refs[v[-1][0]]
        start = v[-1][1]
        end = v[-1][2]
        orient = v[-1][3]
        upstream = ''
        downstream = ''
        if orient == '+':
            if start - 5000 < 0:
                upstream = ref[:start]
            else:
                upstream = ref[start - 5000:start]
                
            if end + 5000 > len(ref):
                downstream = ref[end:]
            else:
                downstream = ref[end:end + 5000]
        else:
            if end + 5000 > len(ref):
                upstream = revcomp(ref[end:])
            else:
                upstream = revcomp(ref[end:end + 5000])
            
            if start - 5000 < 0:
                downstream = revcomp(ref[:start])
            else:
                downstream = revcomp(ref[start - 5000:start])
                
        upOut += '>' + k + '|' + v[-2] + '\t' + str(len(upstream)) + '\n' + upstream + '\n'
        downOut += '>' + k + '|' + v[-2] + '\t' + str(len(downstream)) + '\n' + downstream + '\n'
```
